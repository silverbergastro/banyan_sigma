```python
#Read in csv file with data. Data needs to be ordered as Name, RA, DEC, pmra, pmdec, err_pmra, err_pmdec, rv, err_rv, plx, err_plx
df = pd.read_csv('DD_2018-10-19T2304.csv',low_memory=False)

#Get numpy array of values
data = df.values

#Get number of objects
num_objs = data.shape[0]
print(num_objs)

#Open file to write outputs to
f = open('outputfile_diskdetective_gaia_parttwo.csv','w')

#Headers
f.write('Name,118TAU,ABDMG,BPMG,CAR,CARN,CBER,COL,CRA,EPSC,ETAC,HYA,IC2391,IC2602,LCC,OCT,PL8,PLE,ROPH,TAU,THA,THOR,TWA,UCL,UCRA,UMA,USCO,XFOR,FIELD\n')

#For gauging how fast it's running
start_time = time.time()

numvec = []
timevec = []

#For each object
for i in range(num_objs):
    #Extract data for an object
    data_row = data[i,:]
    
    #Start converting to what the program needs
    data_row_use = [str(data_row[0])]
    
    #For each entry in the data row
    for ent in data_row[1:]:
        #Convert blank values to NaNs
        if ent == 'None':
            data_row_use.append(np.nan)
        else:
            data_row_use.append(ent)
            
    #Eliminate bad data for parallax
    if data_row[9] < 0.:
        data_row_use[9] = np.nan
        data_row_use[10] = np.nan
    #print(data_row_use)
    
    #Run Banyan Sigma
    output_res = banyan_sigma_wrapper(str(data_row_use[0]),None,data_row_use[1],data_row_use[2],data_row_use[3],data_row_use[4],data_row_use[5],data_row_use[6],data_row_use[7],data_row_use[8],data_row_use[9],data_row_use[10])
    #print(output_res)
    
    #Take outputs and convert to numpy array
    output_data = output_res.values[0]
    
    #Convert outputs to a string
    output_string = str(data_row[0])
    for entry in output_data:
        output_string = output_string + ',' + str(entry)
    
    #Write string to file
    f.write(output_string + '\n')
    end_time = time.time()
    #Print how we're doing speed-wise
    print(i+1,time.time()-start_time, (end_time - start_time)/float(i+1), ((end_time - start_time)/float(i+1))*float(num_objs-i-1),((end_time - start_time)/float(i+1))*float(num_objs-i-1)/60.)
    numvec.append(i+1)
    timevec.append((end_time - start_time)/float(i+1))
#Close file
f.close()

print('Done')

plt.figure()
plt.plot(numvec, timevec)
plt.show()

#test = np.zeros((2,3))

#print(test.shape[0])
```
